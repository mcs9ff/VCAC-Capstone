```python
import pandas as pd
import numpy as np

def clean_dataframe(df):
    # Convert everything to string
    df = df.astype(str)

    # Remove unwanted substrings
    df.replace(['.00', ',', '$\t'], '', regex=True, inplace=True)

    # Convert to integer or NaN
    df = df.apply(pd.to_numeric, errors='coerce')

    return df

def align_columns(dataframes):
    # Find common columns
    common_columns = set(dataframes[0].columns)
    for df in dataframes[1:]:
        common_columns.intersection_update(df.columns)

    # Identify columns that are not present in all dataframes
    missing_columns = []
    for i, df in enumerate(dataframes):
        missing_in_df = set(df.columns) - common_columns
        if missing_in_df:
            missing_columns.append((i, list(missing_in_df)))

    # Clean dataframes to only include common columns
    cleaned_dataframes = [df[list(common_columns)] for df in dataframes]

    return cleaned_dataframes, missing_columns
```

```python
def clean_column_names(df):
    # Replace two spaces with one in column names
    df.columns = df.columns.str.replace('  ', ' ')
    df.columns = df.columns.str.replace('$', 'dollars')
    df.columns = df.columns.str.replace(' ', '_')
    df.columns = df.columns.str.replace(':', '_')
    df.columns = df.columns.str.replace('/', '_')
    df.columns = df.columns.str.replace('__', '_')
    df.columns = df.columns.str.replace('+', '_plus')
    df.columns = df.columns.str.lower()
    df.columns = df.columns.str.strip()
    return df

df2015 = clean_column_names(df2015)
df2016 = clean_column_names(df2016)
df2017 = clean_column_names(df2017)
df2018 = clean_column_names(df2018)
df2019 = clean_column_names(df2019)
df2020 = clean_column_names(df2020)
```

```python
# geocode the schools
import requests
import json
import time
import geocoder
import pandas as pd

school_errors_refine = {"George Washington High School(UVA)": "701 Broad St, Danville, VA 24541", 
"Lancaster High School (UVA)": "8815 Mary Ball Rd, Lancaster, VA 22503",
"King & Queen Central High School": "17024 The Trail, King and Queen Court House, VA 23085",
"Eastside High School": "314 Schoolhouse Hill Dr NE, Coeburn, VA 24230", 
"Cumberland County High School": " 15 School Rd, Cumberland, VA 23040",
"Union High School (UVA)": "Big Stone Gap, VA 24219",
"T.C. Williams High School": "Alexandria City High School",
"Nelson County High School":"Nelson High School"}

def geocode_schools(school_names):
    df = pd.DataFrame()
    school_errors = []
    for s in school_names:
        try:
            g = geocoder.osm(s+", VA")
            s_df = pd.json_normalize(g.json)
            s_df['school_name'] = s
            df = pd.concat([df, s_df], axis=0)
            print(s)
        except:
            try:
                g = geocoder.osm(school_errors_refine[s]+", VA")
                s_df = pd.json_normalize(g.json)
                s_df['school_name'] = s
                df = pd.concat([df, s_df], axis=0)
                print(s)
                continue
            except: 
                print("error for ", s)
                school_errors.append(s)
                continue
    return df, school_errors

school_df, school_errors = geocode_schools(school_names)

# save the school_df
school_df.to_csv("school_df.csv")

# load the school_df
school_df = pd.read_csv("school_df.csv")
school_df.drop(columns=['Unnamed: 0'], inplace=True)
school_df.sort_values(by=['school_name'])
school_df.reset_index(drop=True, inplace=True)
school_df.reset_index(drop=False, inplace=True)
keep_cols = ['index', 'address', 'lat', 'lng', 'school_name', 'county', 'state', 'postal', 'raw.boundingbox']
school_df = school_df[keep_cols]

display(school_df)
```
