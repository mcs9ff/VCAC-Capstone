```python
def align_columns(dataframes):
    # Find common columns
    common_columns = set(dataframes[0].columns)
    for df in dataframes[1:]:
        common_columns.intersection_update(df.columns)

    # Identify columns that are not present in all dataframes
    missing_columns = []
    for i, df in enumerate(dataframes):
        missing_in_df = set(df.columns) - common_columns
        if missing_in_df:
            missing_columns.append((i, list(missing_in_df)))

    # Clean dataframes to only include common columns
    cleaned_dataframes = [df[list(common_columns)] for df in dataframes]

    return cleaned_dataframes, missing_columns
```

```python
def clean_column_names(df):
    # Replace two spaces with one in column names
    df.columns = df.columns.str.replace('  ', ' ')
    df.columns = df.columns.str.replace('$', 'dollars')
    df.columns = df.columns.str.replace(' ', '_')
    df.columns = df.columns.str.replace(':', '_')
    df.columns = df.columns.str.replace('/', '_')
    df.columns = df.columns.str.replace('__', '_')
    df.columns = df.columns.str.replace('+', '_plus')
    df.columns = df.columns.str.lower()
    df.columns = df.columns.str.strip()
    return df

df2015 = clean_column_names(df2015)
df2016 = clean_column_names(df2016)
df2017 = clean_column_names(df2017)
df2018 = clean_column_names(df2018)
df2019 = clean_column_names(df2019)
df2020 = clean_column_names(df2020)
```

```python
# bring back the school names
df_imputed['school_name'] = schools

# convert the year_starting to a string
df_imputed['year_starting'] = df_imputed['year_starting']
```

```python
# convert from counts to rates per student
def convert_to_rates(df):
    col_names = list(df.columns)
    rate_col_names = []
    cols = [] 
    for col in col_names:
        cols.append(col)
        rate_col_names.append(col + "_rate")
        
    rate_col_names.remove('student_totals_seniors_rate')
    rate_col_names.remove('school_name_rate')
    rate_col_names.remove('year_starting_rate')
    cols.remove('student_totals_seniors')
    cols.remove('school_name')
    cols.remove('year_starting')

    # # for all the columns divide the original dataframe value by the total seniors value
    for rate_name, col_name in zip(rate_col_names, cols):
        df[rate_name] = df[col_name]/df['student_totals_seniors']
    return df

df = convert_to_rates(df_imputed)

df.to_csv('cleaned_combined_2015_2020.csv', index=False)
```
