```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('cleaned_combined_2015_2020.csv')
```

```python
df.head()
```

```python
schools_grouped = df.groupby('school_name')

plt.figure(figsize=(22, 6)) 
for name, group in schools_grouped:
    plt.plot(group['year_starting'], group['enrolled_rate'], label=name)

plt.xlabel('Year')
plt.ylabel('Enrollment Rate')
plt.title('Enrollment Rates Over Years')
#plt.legend()
plt.grid(True)
plt.show()
```
