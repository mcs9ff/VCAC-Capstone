```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

def create_pairs_plot(df, variables):
    g = sns.PairGrid(df[variables])
    g.map_lower(sns.scatterplot)
    g.map_upper(corr_func)
    g.map_diag(sns.histplot)
    plt.show()

def corr_func(x, y, **kwargs):
    r = np.corrcoef(x, y)[0][1]
    ax = plt.gca()
    ax.annotate(f'Corr: {r:.2f}', xy=(.1, .9), xycoords=ax.transAxes)
```

```python
create_pairs_plot(df, ['fafsa_submitted_rate', 'institutional_aid_total_dollars_rate', 'specific_interactions_1-on-1s_total_rate', 'college_waivers_total_#_waivers_rate', 'student_totals_seniors', 'year_starting', 'school_name'])
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import zscore

def plot_scatter(df, x_var, y_var):
    # Calculate Z-scores
    df['Z_x'] = np.abs(zscore(df[x_var]))
    df['Z_y'] = np.abs(zscore(df[y_var]))

    # Create a scatter plot with a regression line
    sns.regplot(x=x_var, y=y_var, data=df, color='blue', line_kws={'color': 'black'})

    # Highlight outliers
    outliers = df[(df['Z_x'] > 3) | (df['Z_y'] > 3)]
    plt.scatter(outliers[x_var], outliers[y_var], color='red', label='Outliers')

    # Set labels and title
    plt.xlabel(x_var)
    plt.ylabel(y_var)
    plt.title(f'{y_var} vs {x_var}')

    # Show the plot
    plt.show()

    # Print the correlation
    print("Correlation: ", df[x_var].corr(df[y_var]))
```

```python
def create_boxplot(df, x_var, y_var):
    """Create a boxplot of y_var by x_var.

    Args:
        df (_type_): _description_
        x_var (_type_): _description_
        y_var (_type_): _description_
    """
    sns.boxplot(x=x_var, y=y_var, data=df, palette='Set1')
    plt.title(f'{y_var} by {x_var}')
    plt.xlabel(x_var)
    plt.ylabel(y_var)
    plt.xticks(rotation=45)
    plt.show()
```

```python
create_boxplot(df, 'year_starting', 'enrolled_rate')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

def create_lineplot(df, x_var, y_var, group_var):
    """Create a line plot of y_var by x_var, grouped by group_var.

    Args:
        df (pandas.DataFrame): The DataFrame containing the data.
        x_var (str): The column name in df for the x variable.
        y_var (str): The column name in df for the y variable.
        group_var (str): The column name in df for the grouping variable.
    """
    sns.lineplot(x=x_var, y=y_var, hue=group_var, data=df)
    plt.title(f'{y_var} by {x_var}, grouped by {group_var}')
    plt.xlabel(x_var)
    plt.ylabel(y_var)
    plt.xticks(rotation=45)
    plt.legend(title=group_var, bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
    plt.show()
```

```python
def outliers(df, x_var, y_var):
    """Find the outliers in a column of a DataFrame.

    Args:
        df (pandas.DataFrame): The DataFrame containing the data.
        column (str): The column name in df to find outliers in.

    Returns:
        pandas.DataFrame: A DataFrame containing the outliers.
    """
    df['Z_x'] = np.abs(zscore(df[x_var]))
    df['Z_y'] = np.abs(zscore(df[y_var]))

    # Highlight outliers
    outliers = df[(df['Z_x'] > 3) | (df['Z_y'] > 3)]
    return outliers
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import zscore

def plot_scatter(df, x_var, y_var, outliers=False):
    # Calculate Z-scores
    df['Z_x'] = np.abs(zscore(df[x_var]))
    df['Z_y'] = np.abs(zscore(df[y_var]))

    # If outliers is True, remove the outliers
    if outliers:
        df = df[(df['Z_x'] <= 3) & (df['Z_y'] <= 3)]

    # Create a scatter plot with a regression line
    sns.regplot(x=x_var, y=y_var, data=df, color='blue', line_kws={'color': 'black'})

    # Highlight outliers
    outliers = df[(df['Z_x'] > 3) | (df['Z_y'] > 3)]
    plt.scatter(outliers[x_var], outliers[y_var], color='red', label='Outliers')

    # Set labels and title
    plt.xlabel(x_var)
    plt.ylabel(y_var)
    plt.title(f'{y_var} vs {x_var}')

    # Show the plot
    plt.show()

    # Print the correlation
    print("Correlation: ", df[x_var].corr(df[y_var]))
```

```python
plt.style.use('bmh')
plot_scatter(df, 'econ_total_rate', 'enrolled_rate', outliers=True)
```

```python
plt.style.use('bmh')
plot_scatter(df, 'econ_enrolled_rate', 'enrolled_rate', outliers=True)
```

```python
import matplotlib.pyplot as plt

# fivethirtyeight style
plt.style.use('bmh')

# Assuming df is your DataFrame and it has columns 'Senior_Students' and 'VDOE_Students'
total_senior_students = df['student_totals_seniors'].sum()
total_vdoe_students = df['vdoe_total'].sum()

categories = ['Senior Students', 'VDOE Students']
totals = [total_senior_students, total_vdoe_students]

# inlcude text on the bars for the exact number
for i in range(len(categories)):
    plt.text(i, totals[i] + 300, int(totals[i]), ha='center')

plt.bar(categories, totals, color=['blue', 'green'])
plt.title('Comparison of Total Senior Students and VDOE Students')
plt.ylabel('Total Students')
plt.show()
```

There is a difference in the set of students reported as seniors in the VCAC data and the set of students reported on in the VDOE data for the same schools and years. There is not a precise way to trace back ascertain the cause of the inconsistency, however we could presume this may be attributed to differences in the reporting times between observation of data reported in the VCAC data and the enrollement data reported by and through VDOE. We mitigate this by exploring relationships on a per student rate basis.

```python
plt.style.use('bmh')
y_var = 'school_name'
x_var = 'enrolled_rate'
sns.boxplot(x=x_var, y=y_var, data=df, palette='Set1')
plt.axvline(df[x_var].median(), color='black', linestyle='--')
plt.title('Neighborhoods Matter: Enrolled Rate by School')
plt.xlabel(x_var)
plt.ylabel(y_var)
plt.xticks(rotation=45)
plt.show()
```

Within the VCAC/VDOE data, there are 23 schools with useable information over the reviewed time period. The average enrollment rate across across all schools and time periods is 0.60, meaning that ~60% of students in the program enroll. Of the 23 schools with useable data, there appears to be substantial variation in the enrollment rate across the time period. Enrollment rate spreads at a particular school over the time period reviewed vary from school to school. Additionally the enrollment rates between schools appear to vary as well. The school could potentially be capturing or representing other, currently unobserved, factors that could be influencing the enrollment rate. Any conclusions that have not exhaustively considered the factors that make these schools different could over-confidently misrepresent findings due to the potential for omitted variable bias.

It is possible there is a school or district or county effect, but it is also possible that without looking at individual level student data we can not definitively say more than there is a difference. The determination of whether a potential significant difference is caused by a single variable or if it is caused by other external unobserved factors, interventions, or policies is not possible without further data.
